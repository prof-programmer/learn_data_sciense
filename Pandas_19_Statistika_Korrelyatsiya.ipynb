Korrelyasiya yordamida 2 ta ma'lumot o'xshashligini tekshirsa bo'ladi

```python
data = {
    'age':[10,20,30,40,50,60],
    'year':[2010,2000,1990,1980,1970,1960]
}

obj = pd.DataFrame(data)
obj
```

```python
obj['age'].corr(obj['year'])
# bu degani yoshning yilga bog'liqligi, yaqiqatdan yosh yilga bog'liq, ya'ni 100% korsatmoqda
```
