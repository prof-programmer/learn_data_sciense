```python
import pandas as pd
import numpy as np
```

**Bir nechta qatorni tashlab yuborsa ham boladi**

*Qatorlarni list korinishida kiritamiz*

```python
df.drop([20, 17, 15]) # bir nechta qator tashlangan DataFrame ni qaytaradi
```

E'tibor qiling asosiy DataFrame o'zgarishsiz qoladi.

Tashlab yuborilgan holda saqlash uchun uni o'zgaruvchiga saqalash kerak

```python
df
```

**Ustunni tashlab yuborish**

```python
df.drop('Yil', axis=1)
```

**Parametrga axis=1 ni o'rniga axis='columns' deb ham kiritish mumkin**

```python
df.drop('Yil', axis='columns')
```

```python
# Yangi DataFrame ga saqlash
df_new = df.drop('Yil', axis='columns')
```

```python
df_new
```

**O'zgaruvchiga saqlamasdan turib ham DataFrame ni mavjud ishlatilayotgan DataFrame ga saqlasa bo'ladi. **

Buning uchun drop() parametriga qo'shimcha inplace=True parametrini qo'shish kerak
Bu parametrni ishlatishda ehtiyot bolish kerak, sababi siz ishlatib turgan asl DataFramega o'zgartirish kiritib yuboradi va ma'lumotingizni yo'qotishingiz mumkin.

```python
df.drop('Yil', axis='columns', inplace=True)
df
```

inplace parametrni indekslarda ham qollasa bo'ladi

```python
df.drop([10, 15], inplace=True)
df
# SHunday qilib bizda 10, 15 qatorlar tushib qoldi
```
