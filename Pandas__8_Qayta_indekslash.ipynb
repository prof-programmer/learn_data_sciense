```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Yil' : [2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
    'Aholi soni' : [33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
    'Temp' : [1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53],
    'Zichlik' : [79.77, 78.68, 77.53, 76.34, 75.13, 73.91, 72.71, 67.03]
}
df = pd.DataFrame(data)
df
```

```python
df.index = [21, 20, 19, 18, 17, 16, 15, 10]
```

```python
df
```

```python
# Reindex
# indekslar boyicha ma'lumot ketma-ketligini o'zgartirish

ind = [21, 20, 19, 18, 17, 16, 15, 10] # DataFrame da nechta indeks bolsa oshancha sondagi indeks lari kiritildi
ind.sort() # index listi sortirovka qilindi, o'sib borish tartibida
ind
```

```python
df.reindex(ind)
```

```python
df # DataFrame ni chaqirsak, ma'lumotni o'zi tartibi o'zgarmagan
# chunki reindex() ni o'zi yangi dataframe qaytaradi, uni o'zgaruvchiga saqlashimiz kerak
```

```python
df2 = df.reindex(ind)
```

```python
df2 # ana endi indekslar boyicha sortirovka qilingan DataFrame oldik
```

```python
df.reindex([10,15,20]) # reindex()ga bizga kerakli indekslarni list korinishida
# bersak ham boladi, aynan osha ma'lumotlarni bizga qaytaradi
```

```python
df.reindex(index=range(10,22)) # 10 dan 22 gacha bolgan indekslarni chiqarish
# lekin 11-14 indekslarda ma'lumot yoqligi uchun NaN qiymat beriladi
```

```python
# NaN ni o'rniga ma'lumotlar toldirilgan DataFrame olish uchun uni eng yaqin
# qiymatlar bilan toldirish mumkin
df.reindex(index=range(10,22), method='nearest')
```

```python
# metodlar kop: bosh ma'lumotni oldingi yoki keyingi qatordagi ma'lumot bilan to'ldirib qo'ysa bo'ladi
df.reindex(index=range(10,22), method='ffill')
```

```python
# Bizga kerakli ustunlarni qoldirish
df3 = df.reindex(columns=['Aholi soni', 'Temp', 'Zichlik'])
df3.head()
```

```python
df3.sample() # tasodifiy qatorni qaytaradi
```

```python
df3.sample(3) # tasodifiy 3 ta qatorni qaytaradi, har gal tasodifiy qatorlarni qaytaradi
```

```python
# Bizga kerakli indeks va ustunlarni bitta metod bilan chiqarib olamiz
df.loc[[10,15,18,20],['Aholi soni', 'Temp', 'Zichlik']]
```
