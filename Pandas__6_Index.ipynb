```python
df = pd.DataFrame(data, columns=['Yil', 'Temp', 'Aholi soni', 'YAIM'])
```

```python
df
```

```python
df.YAIM = 0 # YAIM ustunni parametr sifatida ko'rsatib qiymatini o'zgartiramiz
```

```python
df
```

YAIM ustunida ba'zi qiymatlar qo'shamiz

```python
val = pd.Series([1700, 1650, 1300], index=[0, 2, 7]) #Series obyektini yaratdik
df.YAIM = val
```

```python
df
```

Endi barcha qiymatlarni o'zgartiraylik
Buning uchun ro'yxat uzunligi qiymatlar soniga teng bo'lishi kerak.
Bizda 8 ta qiymat bor, shu sababli 8 ta qiymatli ro'yxatni ko'rsatishimiz kerak.

```python
val2 = [1700, 1650, 1600, 1590, 1500, 1460, 1400, 1300]
df.YAIM = val2
df['AYIM'] = val2 # ustunni ro'yxat sifatida ko'rsatsak ham bo'ladi, bu ikkinchi variant
```

```python
df
```

Keling yangi '1500+' nomli ustun qo'shamiz va agar YAIM ning qiymati 1500 dan katta bo'lsa shu '1500+' ustunda True qiymat bersin

```python
df['1500+'] = df.YAIM>1500 # bu yerda xuddi lug'atdagidek ishladik
```

```python
df
```

Lug'atdagidek ishladik, xuddi lug'atdagidek ustunni o'chirsak ham bo'ladi

```python
del df['1500+'] # ustunni lug'atdagi kabi o'chiramiz
```

```python
df
```
