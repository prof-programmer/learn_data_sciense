```python
import pandas as pd
import numpy as np
```

```python
s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=['a', 'c', 'd', 'e'])
s2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1], index=['a', 'c', 'e', 'f', 'g'])
```

Ikki Series yoki DataFrame obyektlari o'rtasida arifmetik amallar bajarishda indekslar juda muhim rol o'ynaydi.

Amallar mos tushuvchi indekslar o'rtasida bajariladi. Agar obyektlarning birida mavjud indeks ikkinchisda yo'q bo'lsa natija NaN bo'ladi.

```python
s1
```

```python
s2
```

```python
s1+s2
```

E'tibor bering, s1 da mavjud d indeksi s2da mavjud bo'lmaganligi sabab s1+s2 da d indeksdagi qiymat NaN bo'ldi.

Yuqoridagi holat DataFramelarga ham xos. Serieslardan farqli ravishda DataFrameda qator va ustun indekslari hisobga olinadi.

```python
df1 = pd.DataFrame(np.arange(9.).reshape(3,3), columns=list('bcd'), index=['Olma','Anor','Uzum'])
df2 = pd.DataFrame(np.arange(12.).reshape(4,3), columns=list('abc'), index=['Olma','Anor','Qovun','Anjir'])
```

```python
df1
```

```python
df2
```

```python
df1+df2
```

a va d ustunlar hamda Qovun va Anjir qatorlar ikkala DataFrameda bo'lmagani sababli yuqoridagi natija chiqdi

Yuqoridagi qo'shish amalini .add() metodi yordamida ham bajarish mumkin.

```python
df1.add(df2)
```

Bunda maxsus `fill_value` parametri yordamida mos tushuvchi indekslardagi NaN qiymatlarni boshqa qiymat bilan almashtirish mumkin. Bu usulda yakuniy natijada ikkala df da ham mavjud bo'lmagan indekslarda `NaN` qiymati qoladi.

```python
df2.add(df1, fill_value=0)
```

```python
df1
```

```python
df2
```

### Boshqa arifmetik amallar

- `add`, `radd` - Qo'shish (+)
- `sub`, `rsub` - Ayirish (-)
- `div`, `rdiv` - Bo'lish (/)
- `floordiv`, `rfloordiv` - Bo'lish va butun qismini olis (//)
- `mul`, `rmul` - Ko'paytirish (*)
- `pow`, `rpow` - Eksponenta (\*\*)

Yuqoridagi amallarning barchasida `fill_value` parametri mavjud.

`r`bilan boshlangan metodlarda `DataFrame` ifodaning o'ng  tarafida bo'ladi.

```python
df1
```

```python
df1.sub(2) # df-2 degani
```

```python
df2.rsub(2) # 2-df degani
```

## pandas Obyekti elementlariga Funksiyalarni qo'llash

```python
df = pd.DataFrame(np.random.randn(3,4), index=list('abc'), columns=['Olma','Anor','Qovun','Anjir'])
df
```

Biror funksiyani DF barcha elementlariga qo'llash:

```python
np.abs(df)
```

```python
np.round(df)
```

```python
df.mean()
```

Yuqoridagi funksiyalarni `.apply()` metodi yordamida ham qo'llash mumkin.

```python
df.apply(np.mean)
```

`.apply()` yordamida funksiyani ustun yoki qatorga qo'llashni ko'rsatishimiz mumkin

```python
df.apply(np.sum, axis="columns")
```

Bu metodning afzalligi, o'zingiz istalgan funksiyani yozib uni DF elementlariga qo'llashigniz mumkin:

```python
# DF qatori max va min qiymati ayirmasini hisoblaydigan funksiya
f = lambda x: x.max() - x.min()
```

```python
df.apply(f)
```

Funksiya nafaqat yagona qiymat, balki series obyektini ham qaytarishi mumkin

```python
def foo(x):
    return pd.Series([x.min(), x.max()], index=['min', 'max'])
```

```python
df.apply(foo)
```

Biror funksiyani DF har bir elementiga alohida qo'llash uchun `.applymap()` metodidan foydalaning.

```python
f = lambda x : x*x # qiymatni kvadratga oshiruvchi funksiya
```

```python
df = pd.DataFrame(np.arange(1,13).reshape(3,4), index=list('abc'), columns=['Olma','Anor','Qovun','Anjir'])
df
```

```python
df.applymap(f)
```

`DataFrame`ni qatorlari yoki ustunlari bo'yicha tartilash mumkin

```python
# Keling keyingi mashqlar uchun Githubdan kattaroq DF yuklab olamiz
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv', index_col=0)
df.head()
```

```python
df.sample()
```

```python
df.sort_index()
```

```python
df.sort_index(axis=1) # ustun nomi bo'yicha tartiblash
```

```python
df.sort_index(axis=1, ascending=False) # teskari tartiblash
```

`DataFrame` elementlarini biror ustun(lar) bo'yicha tartiblash ham mumkin

```python
df.sort_values(by='price', ascending=False)
```

```python
df.sort_values(by=['price','mileage'], ascending=False)
```

`.rank()` metodi yordamida Df qatorlarini reytingini aniqlash mumkin. Reyting aniqlashda aynan qaysi ko'rsatkich (ustun) bo'yicvha reytingni aniqlash ko'rsatiladi

```python
# Moshinalarni mileage (yurgan kilometri) bo'yicha reyitnglaymiz
df['rank'] = df['mileage'].rank()
df
```

```python
df.sort_values('mileage')
```

```python
# .rank() method parametri yordamida reyting o'lchash usulini tanlash mumkin (min, max, average, first, dense)
df['rank'] = df['year'].rank(method='average')
df.sort_values('year')
```

```python
df['rank'] = df['price'].rank(method='min')
df.sort_values('price')
```

```python
df['rank'] = df['color'].rank(method='average')
df.sort_values('color')
```
